```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

Rel plots

```python
tips = sns.load_dataset("tips")
tips
```

```python
# scatter plot

# 두개의 연속형 변수 total_bill과 tip간의 관계를 scatter plot으로 표현할수 있습니다

sns.relplot(data=tips, x="total_bill", y="tip")
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
# 크게 hue, size, style을 통해서 데이터포인트가 가지는 categorical 변수값을 표현할 수 있습니다

sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="time",size="size"
)
```

Dis plots

```python
# histogram

penguins = sns.load_dataset("penguins")
sns.histplot(penguins, x="flipper_length_mm")
sns.displot(penguins, x="flipper_length_mm")
```

```python
sns.displot(penguins, x="flipper_length_mm", binwidth=3)
```

```python
# histogram의 막대 너비는 binwidth 혹은 bins를 사용해 조절해줄 수 있습니다
sns.displot(penguins, x="flipper_length_mm", bins=20)
```

```python
# 변수가 이산형 변수인 경우, discrete=True 옵션을 통해 x축 값 중앙에 막대가 위치하도록 할 수 있습니다

sns.displot(tips, x="size", discrete=True)
```

```python
# 또한, shrink 옵션을 통해 막대 간 공간 조절이 가능합니다
sns.displot(tips, x="size", discrete=True, shrink=.7)
```

```python
# displot의 경우 hue 옵션을 통해 추가 변수를 조건으로 한 분포를 시각화 할 수 있습니다
sns.displot(penguins, x="flipper_length_mm", hue="species")
```

```python
# 막대 간 겹쳐지는 부분의 경우 element=step 옵션을 사용하거나 multiple={stack, dodge} 옵션을 사용 가능합니다
sns.displot(penguins, x="flipper_length_mm", hue="species", element="step")
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
sns.displot(tips, x="size", discrete=True, shrink=.7, hue='sex', multiple='dodge')
```

```python
# distplot 또한 col이나 row를 이용해 변수에 따른 여러 플롯으로 나눠 그릴 수 있습니다
sns.displot(penguins, x="flipper_length_mm", col="sex")
```

```python
# stats='probability' 옵션을 통해 모든 막대의 높이의 합이 1이 되게 만들 수 있으며, 이를 빈도의 비율값으로 해석할 수 있습니다.
sns.displot(penguins, x="flipper_length_mm", hue="species", stat="probability")
```

```python
# kde plot을 이용해 histogram과 다른 부드러운 곡선으로 이어진 분포 시각화가 가능합니다
sns.kdeplot(penguins, x="flipper_length_mm")
```

```python
# bw_adjust를 이용해 곡선을 세밀하게 혹은 더 부드럽게 표현이 가능합니다
sns.displot(penguins, x="flipper_length_mm", kind="kde", bw_adjust=.25)
```

```python
# kde plot 또한 hue를 이용해서 변수간 분포를 더 세밀하게 시각화할 수 있습니다
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="kde")
```

```python
# 또한, fill=True 옵션을 이용해 곡선 아래 색칠된 형태로 표현이 가능합니다
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="kde", fill=True)
```

```python
# hue 옵션을 사용하는 대신, y에 2번째 변수를 넣어 여러 변수의 분포를 표현하는 시각화 또한 가능합니다
# 기본적으로 displot은 heatmap을 그려줍니다
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
# kde plot에서 y 변수를 추가해주면 등고선 플롯을 그려줍니다
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", kind="kde")
```

```python
# heatmap과 contour plot 또한 hue 옵션을 추가해서 hue 변수에 따라 여러 분포를 구별해서 볼 수 있습니다
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="kde")
```

```python
# heatmap의 경우 (x축,y축)의 binwidth를 조절하여 각 칸의 크기를 조절해줄 수 있습니다.
# 또한, cbar=True 로 설정시 색깔에 따라 갖는 빈도 값을 확인할 수 있습니다

sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", binwidth=(2, .5), cbar=True)
```

```python
# 등고선 플롯의 경우 thresh와 levels를 이용해 선의 간격을 조절할 수 있습니다
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", kind="kde", thresh=.2, levels=4)
```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", kind="kde", levels=[.01, .05, .1, .8])
```

```python
# joint plot의 경우 2개의 변수 간 분포와 각 변수의 분포를 동시에 보여줍니다
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
# kind=’kde’ 인 경우 contour plot과 kde plot을 그려줍니다
sns.jointplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="species",
    kind="kde"
)
```

```python
# Pair plot을 이용해 각 변수들간의 pairwise 관계를 시각화할 수 있습니다
# 특히, 고려해야할 변수들이 많은데 변수간 상관관계 파악이 필요한 경우 유용합니다
sns.pairplot(penguins)
```

Catplot

```python
tips = sns.load_dataset("tips")
tips
```

```python
# catplot의 기본 플롯은 strip plot이며, 한쪽 변수가 categorical 데이터인 산점도라고 생각하면 됩니다
# 한 범주에 속하는 모든 점이 범주형 변수에 해당하는 축을 따라 동일한 위치에 있게 됩니다

sns.catplot(data=tips, x="day", y="total_bill")
```

```python
# jitter 옵션을 통해서 좌우로 퍼진 정도를 조절하거나 비활성화합니다
sns.catplot(data=tips, x="day", y="total_bill", jitter=.3)
```

```python
sns.catplot(data=tips, x="day", y="total_bill", jitter=False)
```

```python
# swarm plot을 활용하면 점들의 중첩을 방지하는 알고리즘을 사용하여 시각화가 가능합니다
# 상대적으로 데이터 수가 적은 경우에만 잘 작동하지만 관측치 분포를 더 잘 표현할 수 있습니다

# sns.swarmplot(data=tips, x="day", y="total_bill")
sns.catplot(data=tips, x="day", y="total_bill", kind="swarm")
```

```python
# 관계형 도표와 마찬가지로 hue를 사용하여 범주형 도표에 다른 차원을 추가할 수 있습니다
# size또는 style는 지원되지 않습니다

sns.catplot(data=tips, x="day", y="total_bill", hue="sex", kind="swarm")
```

```python
# 데이터에 pandas Categorical 데이터 유형이 있는 경우 범주의 기본 순서를 여기에서 설정할 수 있습니다.
tips['day'].dtype
```

```python
# 데이터 유형이 Categorical이 아닌 경우에도 적용 가능하지만 Categorical 데이터로 취급되어 문제가 발생하는 경우가 있습니다.
sns.catplot(data=tips.loc[tips['size']!=3], x="size", y="total_bill")
```

```python
# 이 때, native_scale을 사용하면 x축 데이터의 분포를 보존할 수 있습니다
sns.catplot(data=tips.loc[tips['size']!=3], x="size", y="total_bill", native_scale=True)
```

```python
# order 옵션을 사용해서 Categorical 변수의 순서를 바꿔줄 수도 있습니다
sns.catplot(data=tips, x="smoker", y="tip", order=["No", "Yes"])
```

```python
tips['smoker'].dtypes
```

```python
# x와 y에 들어가는 변수 이름에 따라 축을 가로로 변경 가능합니다
sns.catplot(data=tips, x="total_bill", y="day", hue="time", kind="swarm")
```

```python
# boxplot
# 데이터 세트의 크기가 커짐에 따라 범주형 산점도는 각 범주 내의 값 분포에 대해 제공할 수 있는 정보가 제한됩니다
# 이 때 boxplot을 사용할 수 있습니다

# sns.boxplot(data=tips, x="day", y="total_bill")
sns.catplot(data=tips, x="day", y="total_bill", kind="box")
```

```python
# hue를 추가 가능하며 각 수준에 대한 상자는 더 좁아지고 범주형 축을 따라 이동됩니다.
sns.catplot(data=tips, x="day", y="total_bill", hue="smoker", kind="box")
```

```python
# boxen plot을 이용하면 분위수를 좀 더 자세하게 표시할 수 있습니다
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python

sns.boxenplot(data=diamonds.sort_values("color"), x="color", y="price")
sns.catplot(
    data=diamonds.sort_values("color"),
    x="color", y="price", kind="boxen",
    )
```

```python
# violin plot
# violin plot은 box plot과 kde 곡선을 결합한 형태입니다
# Boxplot에서 쓰인 IQR과 whisker또한 표현됩니다
sns.violinplot(
    data=tips, x="total_bill", y="day", hue="sex",
)
sns.catplot(
    data=tips, x="total_bill", y="day", hue="sex", kind="violin",
)
```

```python
# bw와 cut을 이용해서 kde 곡선을 조절할 수 있습니다
sns.catplot(
    data=tips, x="total_bill", y="day", hue="sex",
    kind="violin", bw_adjust=.5, cut=0,
)
```

```python
# split=True로 설정시 하나의 violin안에 두 변수를 표현할 수 있습니다
sns.catplot(
    data=tips, x="day", y="total_bill", hue="sex",
    kind="violin", split=True,
)
```

```python
# box plot이나 violin plot을 swarm plot과 결합하는 것도 가능합니다
# swarmplot의 ax에 box plot이나 violin plot을 넣어줍니다
# 이 때, violin plot에서 inner=None을 통해 내부의 boxplot을 제거해줍니다
g = sns.catplot(data=tips, x="day", y="total_bill", kind="violin", inner=None)
sns.swarmplot(data=tips, x="day", y="total_bill", color="k", size=3, ax=g.ax)
```

```python
g = sns.catplot(data=tips, x="day", y="total_bill", kind="box")
sns.swarmplot(data=tips, x="day", y="total_bill", color="k", size=3, ax=g.ax)
```

```python
# barplot
# catplot을 통해서도 barplot을 그릴 수 있습니다
# 이 때, 추가적으로 95% 신뢰구간을 볼 수 있습니다
titanic = sns.load_dataset("titanic")
sns.catplot(data=titanic, x="sex", y="survived", hue="class", kind="bar")
```

```python
# errorbar 옵션을 이용해서 다른 방식으로 분산을 표현할 수 있습니다
sns.catplot(data=titanic, x="age", y="deck", errorbar="sd", kind="bar")
```

```python
# Count plot

# Categorical 변수의 빈도를 보기 위해서는 count plot을 사용할 수 있습니다.
sns.catplot(data=titanic, x="deck", kind="count")
```

```python
sns.catplot(
    data=titanic, y="deck", hue="class", kind="count",
    palette="pastel", edgecolor=".6",
)
```

```python
# Pointplot
# Pointplot은 Categorical 변수에 따라 다른 변수가 어떻게 변화하는지를 효과적으로 나타냅니다
# 이 때 점 추정치 중심으로 95% 신뢰구간이 표현됩니다
sns.catplot(data=titanic, x="sex", y="survived", hue="class", kind="point")
```

```python
# marker, linestyles 옵션을 이용해 선과 점 추정치의 형태를 바꿔줄 수 있습니다
sns.catplot(
    data=titanic, x="class", y="survived", hue="sex",
    palette={"male": "g", "female": "m"},
    markers=["^", "o"], linestyles=["-", "--"],
    kind="point"
)
```
